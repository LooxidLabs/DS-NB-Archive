```python
def get_s3_obj_from_path(path):
    s3 = boto3.client('s3')

    # S3 버킷 및 객체 경로 설정
    bucket_name = 'focusmate-public-dev'
    object_key = path

    try:
        response = s3.get_object(Bucket=bucket_name, Key=object_key)
        file_content = response['Body'].read()
        json_data = json.loads(file_content)
        
        selected_keys = ["eegChannel1", "eegChannel2"]
        result = {key: json_data[key] for key in selected_keys}

    except Exception as e:
        print(f"객체 다운로드 및 파싱 중 오류 발생: {str(e)}")
        
    return result
```

```python
import numpy as np

from scipy.signal import welch
from scipy import signal

def adjust_to_score_min_max(target):
    if target > 100:
        return 100
    if target < 0:
        return 0
        
    return target

def eeg_filter(eeg, btype, eeg_sr=250, freq_high=0.5, freq_stop=60, freq_band=[1, 50], eeg_butter_order=3, ):
    f_n = eeg_sr / 2
    b = 1
    a = 1
    if btype == "highpass":
        b, a = signal.butter(
            eeg_butter_order, freq_high / f_n, btype=btype, analog=False
        )
    elif btype == "bandstop":
        b, a = signal.butter(
            eeg_butter_order,
            [(freq_stop - 1) / f_n, (freq_stop + 1) / f_n],
            btype,
            analog=False,
        )
    elif btype == "bandpass":
        b, a = signal.butter(
            eeg_butter_order,
            np.array(freq_band) / f_n,
            btype,
            analog=False,
        )

    print(a, b)
    filtered_eeg = signal.filtfilt(
        b,
        a,
        eeg,
        axis=0,
        padtype="odd",
        padlen=3 * (max(len(b), len(a)) - 1),
    )
    return filtered_eeg


def apply_filter(eeg_raw_data):
    filtered_data = []
    eeg = eeg_raw_data

    eeg_aug = np.append(np.flipud(eeg), eeg, axis=0)
    eeg_aug = np.append(eeg_aug, np.flipud(eeg), axis=0)

    applied_highpass = eeg_filter(eeg_aug, "highpass")
    applied_notch = eeg_filter(applied_highpass, "bandstop")
    applied_bandpass = eeg_filter(applied_notch, "bandpass")

    filt_eeg = applied_bandpass[len(eeg): 2 * len(eeg)]
    return filt_eeg


def epoching(eeg, nperseg):
    epochs = [eeg[i * nperseg: (i + 1) * nperseg] for i in range(len(eeg) // nperseg)]
    return epochs


def psd_norm(psd):
    denom = psd.sum(axis=0, keepdims=True)
    normed = psd / denom
    return normed


def calculate_psd(eeg, nperseg, relative=True):
    # Calculate relative PSD
    MIN_FREQ_INDEX = 1
    MAX_FREQ_INDEX = 55

    # epochs = epoching(eeg, 256)

    f, psd = welch(eeg, fs=250, nperseg=nperseg, axis=0, window='hamming')

    idx_crop = (MIN_FREQ_INDEX <= f) & (f <= MAX_FREQ_INDEX)
    psd = psd[idx_crop]
    f = f[idx_crop]

    if relative:
        psd = psd_norm(psd)

    return f, psd


def extract_sustained_attention(feature):
    sustained_score = 1 / (1 + pow(np.e, -feature + 0.16511562718472886)) * 100

    sa_score = eeg_constants.sa_scaler.transform(sustained_score)

    if sa_score <= 10:
        sa_score = 10
    if sa_score >= 99:
        sa_score = 99

    return sa_score


def extract_brain_capacity(feature):
    capacity_score = 1 / (1 + pow(np.e, -feature + 0.12938886069456165)) * 100
    # with open('capacity_scaler.pkl', 'rb') as f:
    #     cp_scaler = pickle.load(f)

    cp_score = eeg_constants.cp_scaler.transform(capacity_score)

    if cp_score <= 10:
        cp_score = 10
    if cp_score >= 99:
        cp_score = 99

    return cp_score
```
